Task1: Exploring and Visualizing the simple dataset

Step 1: Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

Step 2: Load Dataset

```python
df = sns.load_dataset("iris")
```

Step 3: Display the dataset(.shape,.columns,.head())

```python
print("Dataset_heads: \n",df.head())
```

```python
print("Dataset_colmuns: ", df.columns)
```

```python
print("Dataset_shape: ",df.shape)
```

Step 4:Scatter plot to show relationship b/w variables

```python
print("Figure: ", plt.figure())
# 640 wide and 480 tall
```

```python
plt.figure()
plt.scatter(df['sepal_length'], df['petal_length'])
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.title("SCATTER PLOT: sepal_length vs petal_length")
plt.show()
```

```python
plt.figure()
plt.scatter(df['sepal_width'], df['petal_width'])
plt.xlabel("sepal_width")
plt.ylabel("petal_width")
plt.title("SCATTER PLOT:SEPAL WIDTH VS PETAL WIDTH")
plt.show()
```

Step 5: Histogram to examine data distribution

```python
plt.figure()
plt.hist(df['sepal_width'])
plt.xlabel('sepal_width')
plt.ylabel('Frequency')
plt.title("Histogram:Sepal_width")
plt.show()
```

```python
plt.figure()
plt.hist(df['petal_width'])
plt.xlabel('petal_width')
plt.ylabel('Frequency')
plt.title("Histogram:petal_width")
plt.show()
```

```python
plt.figure()
plt.hist(df['sepal_length'])
plt.xlabel('Sepal_Length')
plt.ylabel('Frequency')
plt.title("Histogram:Sepal_length")
plt.show()
```

```python
plt.figure()
plt.hist(df['petal_length'])
plt.xlabel('petal_length')
plt.ylabel('Frequency')
plt.title("Histogram:petal_length")
plt.show()
```

Step 7: Box plot to detect outliers and spread values

```python
plt.figure()
plt.boxplot(df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
plt.xticks([1,2,3,4],['Sepal length','Sepal width','Petal length','Petal width'])
plt.title('Box plot of iris features')
plt.show()
```
